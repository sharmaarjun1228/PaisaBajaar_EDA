
### Column Descriptions and Usage

1. **ID**:  
   - **Description**: Unique identifier for each record in the dataset.  
   - **Usage**: Not necessary for analysis but can be used for tracking individual records.

2. **Customer_ID**:  
   - **Description**: Unique identifier for each customer.  
   - **Usage**: Useful for grouping records by customer when analyzing patterns over time (e.g., monthly behavior).

3. **Month**:  
   - **Description**: Month corresponding to the record.  
   - **Usage**: Can be used to track changes in customer behavior or financials over time (e.g., seasonal trends in spending or credit utilization).

4. **Name**:  
   - **Description**: Name of the customer.  
   - **Usage**: Not used directly in analysis, as it doesn't contribute to credit score prediction.

5. **Age**:  
   - **Description**: Age of the customer.  
   - **Usage**: Important for analysis, as age might affect credit behavior. For example, younger individuals might have less credit history, impacting their scores.

6. **SSN**:  
   - **Description**: Social Security Number of the customer.  
   - **Usage**: Not necessary for analysis and should be treated as sensitive information, excluded from any model.

7. **Occupation**:  
   - **Description**: The profession of the customer (e.g., Scientist, Teacher).  
   - **Usage**: Useful for segmentation, as certain occupations might correlate with different financial behaviors (e.g., high-income professionals may have higher credit scores).

8. **Annual_Income**:  
   - **Description**: Annual income of the customer.  
   - **Usage**: A crucial factor for creditworthiness. Higher incomes generally lead to better credit scores, as individuals are better positioned to repay loans.

9. **Monthly_Inhand_Salary**:  
   - **Description**: Monthly take-home salary of the customer.  
   - **Usage**: Affects affordability, repayment capacity, and credit utilization ratio, all of which are key to determining credit scores.

10. **Num_Bank_Accounts**:  
    - **Description**: Number of bank accounts owned by the customer.  
    - **Usage**: Could indicate the customers financial complexity or wealth. A higher number of accounts might be linked to better financial management or, conversely, overextension.

11. **Num_Credit_Card**:  
    - **Description**: Number of credit cards owned by the customer.  
    - **Usage**: Affects credit utilization and debt. Customers with more credit cards may have higher utilization rates, impacting their credit score.

12. **Interest_Rate**:  
    - **Description**: Average interest rate on loans or credit cards.  
    - **Usage**: A high interest rate can indicate risky behavior or poor creditworthiness, which could negatively affect the credit score.

13. **Num_of_Loan**:  
    - **Description**: Number of loans taken by the customer.  
    - **Usage**: Used to assess debt burden. Multiple loans can indicate higher risk, but also financial responsibility if managed well.

14. **Type_of_Loan**:  
    - **Description**: The types of loans the customer holds (e.g., Mortgage, Auto Loan, Personal Loan).  
    - **Usage**: Different types of loans might carry different levels of risk. Mortgages are typically seen as safer than high-interest personal loans.

15. **Delay_from_due_date**:  
    - **Description**: Average number of days a customer delays payments.  
    - **Usage**: Directly impacts credit score. Late payments are a negative indicator of creditworthiness.

16. **Num_of_Delayed_Payment**:  
    - **Description**: The number of delayed payments made by the customer.  
    - **Usage**: An important predictor for credit risk. More delayed payments typically lower the credit score.

17. **Changed_Credit_Limit**:  
    - **Description**: Whether or not the customers credit limit has been changed.  
    - **Usage**: Credit limit changes (upward or downward) can indicate a customers relationship with creditors, which affects their credit score.

18. **Num_Credit_Inquiries**:  
    - **Description**: The number of times the customers credit has been checked.  
    - **Usage**: Frequent credit inquiries can signal riskier behavior, negatively impacting the credit score.

19. **Credit_Mix**:  
    - **Description**: Indicates the diversity of credit accounts (e.g., credit cards, mortgages, loans).  
    - **Usage**: A healthy mix of credit types generally improves credit scores.

20. **Outstanding_Debt**:  
    - **Description**: The total amount of debt the customer has not yet paid.  
    - **Usage**: A critical factor in determining creditworthiness. Higher debt levels can lower credit scores.

21. **Credit_Utilization_Ratio**:  
    - **Description**: The ratio of used credit to total available credit.  
    - **Usage**: One of the most important factors for credit score calculation. Lower utilization ratios are better for credit scores.

22. **Credit_History_Age**:  
    - **Description**: The length of the customers credit history in months.  
    - **Usage**: Longer credit histories usually lead to better scores, as they show more financial responsibility over time.

23. **Payment_of_Min_Amount**:  
    - **Description**: Whether the customer only makes minimum payments (Yes/No).  
    - **Usage**: Making only minimum payments can indicate financial distress, negatively affecting the credit score.

24. **Total_EMI_per_month**:  
    - **Description**: Total Equated Monthly Installments (EMIs) paid by the customer.  
    - **Usage**: High EMIs relative to income can suggest over-leverage, affecting the ability to take on more credit.

25. **Amount_invested_monthly**:  
    - **Description**: The amount the customer invests monthly.  
    - **Usage**: Investment behavior may indicate financial stability, which can positively influence creditworthiness.

26. **Payment_Behaviour**:  
    - **Description**: A categorical variable describing the customers payment behavior (e.g., High spent, Low spent).  
    - **Usage**: Provides insight into spending habits and can indicate financial discipline or risk.

27. **Monthly_Balance**:  
    - **Description**: The customers remaining balance at the end of each month.  
    - **Usage**: Helps assess financial liquidity, which is important for determining if the customer can manage credit responsibly.

28. **Credit_Score**:  
    - **Description**: The credit score of the customer (e.g., Good, Poor).  
    - **Usage**: This is the target variable for analysis and prediction. The goal is to understand what factors lead to higher or lower credit scores.

````python
# Define income thresholds for grouping
low_income_threshold = 	7005.930000
middle_income_threshold = 36999.705000

# Create a new column for income groups
df['Income_Group'] = 'Low Income'
df.loc[(df['Annual_Income'] > low_income_threshold) & (df['Annual_Income'] <= middle_income_threshold), 'Income_Group'] = 'Middle Income'
df.loc[df['Annual_Income'] > middle_income_threshold, 'Income_Group'] = 'High Income'

````

````python
df['Income_Group']
````

````python
# Create a box plot to visualize credit scores by income group
# Create a box plot to visualize credit scores by income group
plt.figure(figsize=(10, 6))



plt.boxplot([df[df['Income_Group'] == 'Low Income']['Credit_Score'],
             df[df['Income_Group'] == 'Middle Income']['Credit_Score'],
             df[df['Income_Group'] == 'High Income']['Credit_Score']],
            labels=['Low Income', 'Middle Income', 'High Income'])
plt.xlabel('Income Group')
plt.ylabel('Credit Score')
plt.title('Credit Score by Income Group')
plt.show()
````

````python
# Scatter plot and correlation between Annual Income and Credit Score
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Income_Group', y='Credit_Score', data=df)
plt.title('Annual Income vs Credit Score')
plt.show()

# Spearman Correlation (as Credit Score is ordinal)
from scipy.stats import spearmanr
corr_income, _ = spearmanr(df_cleaned['Annual_Income'], df_cleaned['Credit_Score'])
print(f'Spearman Correlation between Annual Income and Credit Score: {corr_income}')
````
